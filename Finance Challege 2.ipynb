Fraud dection model to track key KPIs for fraud

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import class_weight
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import SMOTE

# Load the dataset
data = pd.read_csv('credit_card_data_da.csv')

# Relevant KPIs for customer segmentation
selected_features = ['Amount', 'FICO Score', 'Has Chip', 'Cards Issued', 'Card', 'Zipcode', 'Use Chip', 'CARD INDEX', 'Total Debt', 'Yearly Income - Person', 'Current Age', 'time_of_day', 'total_debt_personal_income_ratio', 'rolling_charge_off', 'years_since_pin_change', 'rolling_tran_count', 'Num Credit Cards']

# Target
target_variable = 'Is Fraud?'

# Filtering and preprocessing the data
filtered_data = data[selected_features + [target_variable]]
filtered_data = filtered_data.dropna()  # Remove rows with missing values

# Encoding the target variable
label_encoder = LabelEncoder()
filtered_data['Is Fraud?'] = label_encoder.fit_transform(filtered_data['Is Fraud?'])

# Convert 'time_of_day' to binary
filtered_data['Is Morning'] = (filtered_data['time_of_day'] == 'morning').astype(int)

# Separating categorical and numerical features
categorical_features = ['Use Chip', 'CARD INDEX', 'time_of_day', 'Has Chip', 'Cards Issued']
numerical_features = [col for col in selected_features if col not in categorical_features]

# Applying one-hot encoding to categorical features
transformers = [
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(drop='first'), categorical_features)
]

preprocessor = ColumnTransformer(transformers)

# Splitting the data into features and target
X = filtered_data[selected_features]
y = filtered_data[target_variable]

# Applying preprocessing to the features
X = preprocessor.fit_transform(X)

# Addressing class imbalance using SMOTE (Synthetic Minority Over-sampling Technique)
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Splitting the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Calculation of class weights to address class imbalance
class_weights = class_weight.compute_class_weight('balanced', classes=[0, 1], y=y_train)

# Random Forest classifier with adjusted class weights and higher precision
rf_classifier = RandomForestClassifier(n_estimators=100, class_weight={0: class_weights[0], 1: class_weights[1]})

# Training the classifier
rf_classifier.fit(X_train, y_train)

# Making predictions with adjusted threshold (higher precision)
y_prob = rf_classifier.predict_proba(X_test)[:, 1]
threshold = 0.5  # Adjust this threshold to increase precision and recall
y_pred = (y_prob > threshold).astype(int)

# Decoding the labels for the classification report
y_test_decoded = label_encoder.inverse_transform(y_test)
y_pred_decoded = label_encoder.inverse_transform(y_pred)

# Evaluation of the classifier
print(classification_report(y_test_decoded, y_pred_decoded))
```

4) Unsupervised Machine Learning for clustering of fraud alerts into groups (K-means) 

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Loading the data
csv_file = r'C:\Users\brown\Documents\DS4A - Finance Challenge - 2\credit_card_data_da.csv'
ccda_df = pd.read_csv(csv_file)

# Selected Features
features = ['Amount', 'FICO Score', 'Cards Issued', 'Card', 'Zipcode', 'CARD INDEX', 'Total Debt', 'Yearly Income - Person', 'Current Age', 'total_debt_personal_income_ratio', 'rolling_charge_off', 'years_since_pin_change', 'rolling_tran_count', 'Num Credit Cards']

# Segmentation data
segmentation_data = ccda_df[features]

# Handle missing values, replpaccing with mean
segmentation_data = segmentation_data.fillna(segmentation_data.mean())

# Standardization of the data
scaler = StandardScaler()
segmentation_data_scaled = scaler.fit_transform(segmentation_data)

# Optimal number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(segmentation_data_scaled)
    wcss.append(kmeans.inertia_)

# Using the Elbow method to help choose the number of clusters
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()

# Columns for clustering
columns_to_cluster = ['Amount', 'FICO Score', 'Cards Issued', 'Card', 'Zipcode', 'CARD INDEX', 'Total Debt', 'Yearly Income - Person', 'Current Age', 'total_debt_personal_income_ratio', 'rolling_charge_off', 'years_since_pin_change', 'rolling_tran_count', 'Num Credit Cards']

# K-means clustering
k = 4  # Number of clusters
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
ccda_df['Cluster'] = kmeans.fit_predict(segmentation_data_scaled)

# Grouping the data by the 'Cluster' label and calculating the mean for each feature.
cluster_profiles = ccda_df.groupby('Cluster')[columns_to_cluster].mean()

# Displaying the cluster profiles.
print(cluster_profiles)

# Export profiles to an Excel file.
output_file = "cluster_profiles.xlsx"
cluster_profiles.to_excel(output_file)

print(f"Cluster profiles saved to {output_file}")
```
