Here we will get a better look at sales, profits, average price per unit, and units sold.

```python
total_sale = df["Total Sales"].sum()
total_profit = df["Operating Profit"].sum()
avg_price = df["Price per Unit"].mean()
total_units_sold = df["Units Sold"].sum()
```

```python
print("Total Sale of Adidas", total_sale)
print ("Total Profit of Adidas", total_profit )
print("Average price of Adidas", avg_price)
print("total units sold of Adidas", total_units_sold)
```

We will Create a bar chart comparing Average Total Sales and Average Operating Profit for each month.

```python
# Convert 'Invoice Date' to datetime
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])

# Extract the month from the 'Invoice Date' column
df['Month'] = df['Invoice Date'].dt.strftime('%b %Y')

# Group by month and calculate the average for 'Total Sales' and 'Operating Profit' for all cities
monthly_avg = df.groupby('Month').agg({'Total Sales': 'mean', 'Operating Profit': 'mean'}).reset_index()

# Sort the data by month
monthly_avg['Month'] = pd.to_datetime(monthly_avg['Month'], format='%b %Y')
monthly_avg = monthly_avg.sort_values(by='Month')

# Create the bar chart
plt.figure(figsize=(10, 6))
width = 0.4

x = range(len(monthly_avg))
plt.bar(x, monthly_avg['Total Sales'], width=width, label='Average Total Sales', align='center')
plt.bar([i + width for i in x], monthly_avg['Operating Profit'], width=width, label='Average Operating Profit', align='center')

plt.xlabel('Month')
plt.ylabel('Average Amount')
plt.title('Average Total Sales and Operating Profit by Month for All Cities')
plt.xticks([i + width / 2 for i in x], monthly_avg['Month'].dt.strftime('%b %Y'), rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

We will identify the months with the Highest Average Total Sales and Highest Operating Profit.

```python
# Convert "Invoice Date" column to datetime
df["Invoice Date"] = pd.to_datetime(df["Invoice Date"])

# Extract month from the "Invoice Date" and create a new column
df["Month"] = df["Invoice Date"].dt.strftime("%Y-%m")

# Group data by month and calculate average for both columns
monthly_avg = df.groupby("Month").agg({"Total Sales": "mean", "Operating Profit": "mean"})

# Find the month with the highest average total sales
max_total_sales_month = monthly_avg["Total Sales"].idxmax()
max_total_sales_value = monthly_avg["Total Sales"].max()

# Find the month with the highest average operating profit
max_operating_profit_month = monthly_avg["Operating Profit"].idxmax()
max_operating_profit_value = monthly_avg["Operating Profit"].max()

print(f"Month with the Highest Average Total Sales: {max_total_sales_month} (Average: {max_total_sales_value})")
print(f"Month with the Highest Average Operating Profit: {max_operating_profit_month} (Average: {max_operating_profit_value})")
```

```python
#Create Pie chart

# Data for the pie chart
labels = ['Total Sales', 'Total Profit']
sizes = [total_sale, total_profit]
colors = ['skyblue', 'lightgreen']
explode = (0.1, 0)  # Explode the first slice (Total Sales)

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Comparison of Total Sales and Total Profit (April 2020 - September 2021)')

# Add a legend
plt.legend()

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Profit by Retailer

```python
# Sort the DataFrame by "Operating Profit" in descending order
df = df.sort_values(by="Operating Profit", ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df["Retailer"], df["Operating Profit"], color='black')
plt.title('Profit by Retailer')
plt.xlabel('Retailer')
plt.ylabel('Operating Profit')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python

# Filter the DataFrame for rows where 'City' is 'Charleston'
charleston_data = df[df['City'] == 'Charleston'].copy()  # Make a copy to avoid modifying the original DataFrame

# Convert 'Invoice Date' to datetime using .loc
charleston_data.loc[:, 'Invoice Date'] = pd.to_datetime(charleston_data['Invoice Date'])

# Extract the month from the 'Invoice Date' column using .loc
charleston_data.loc[:, 'Month'] = charleston_data['Invoice Date'].dt.strftime('%b %Y')

# Group by month and calculate the average for 'Total Sales' and 'Operating Profit'
monthly_avg = charleston_data.groupby('Month').agg({'Total Sales': 'mean', 'Operating Profit': 'mean'}).reset_index()

# Sort the data by month
monthly_avg['Month'] = pd.to_datetime(monthly_avg['Month'], format='%b %Y')
monthly_avg = monthly_avg.sort_values(by='Month')

# Create the bar chart
plt.figure(figsize=(10, 6))
width = 0.4

x = range(len(monthly_avg))
plt.bar(x, monthly_avg['Total Sales'], width=width, label='Average Total Sales', align='center')
plt.bar([i + width for i in x], monthly_avg['Operating Profit'], width=width, label='Average Operating Profit', align='center')

plt.xlabel('Month')
plt.ylabel('Average Amount')
plt.title('Average Total Sales and Operating Profit in Charleston by Month')
plt.xticks([i + width / 2 for i in x], monthly_avg['Month'].dt.strftime('%b %Y'), rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

Charleston, SC appears to be generating the highest sales relative to the other cities, followed by New York, San Francisco, Miami, and Portland. Importantly, Mens followed by Womens apparel respectively, is driving the most sales with sales peaks observed in the months of July and December.
